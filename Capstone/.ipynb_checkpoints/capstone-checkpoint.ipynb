## <a id='toc1_2_'></a>[**Reading, Understanding, Cleaning, and Transforming the data**](#toc0_)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_excel("C:/Work/Coding/Data_analysis/DA_Projects/TDI/Python/Capstone/Telco_customer_churn.xlsx")
data.head()
```

```python
data.info()
```

```python
data.describe(include='all').transpose()
```

**The following columns would be removed**

- **Count**: It serves no real purpose in the data
- **Country**: There's only one country in the data
- **State**: There's also only one state in the data
- **Zip Code**: It also serves not real purpose in the data
- **Lat Long**: It is a redundant column as there are columns for latitude and longitude already

```python
data.drop(["Count","Country","State","Zip Code", "Lat Long"], axis=1, inplace=True)
```

```python
data.head().transpose()
```

```python
data['Total Charges'] = np.where(data['Total Charges'] == " ", np.nan, data['Total Charges'])
data['Total Charges'] = data['Total Charges'].astype(float)
```

```python
data["Churn Reason"].value_counts()
```

```python
def transform_reason(reason):
    if isinstance(reason, str):
        if "Competitor" in reason:
            return 'Competitor'
        elif ("support" in reason) or ("service provider" in reason):
            return "Support"
        elif ("Product dissatisfaction" in reason) or () or ("Network" in reason) or ("Service dissatisfaction" in reason) or ("Limited" in reason)  or ("Lack" in reason):
            return 'Dissatisfaction'
        elif ("Price" in reason) or ("affordable" in reason) or ("distance" in reason):
            return 'Price'
        elif ("Don't" in reason):
            return 'Unknown'
        else:
            return reason
```

```python
data['Churn Reason'] = data['Churn Reason'].apply(transform_reason)
```

```python
data['Churn Reason'].value_counts()
```

```python
((data['Churn Label'] == 'Yes').astype(int) == (data['Churn Value'] == 1)).sum()
```

```python
churned = data[data['Churn Label'] == "Yes"]
```

The charts above illustrate the distribution of customers and churned customers across the state.

These findings suggest that customer churn is not significantly influenced by geographical location.

#### <a id='toc1_4_1_2_'></a>[**Distribution of Customers and Customer Churn  by Gender**](#toc0_)

```python
agg_data_gender = data.groupby(['Gender']).size().reset_index(name='Total')
agg_churned_gender = churned.groupby(['Gender']).size().reset_index(name='Churned')

churned_rates_gender = pd.merge(left=agg_data_gender, right=agg_churned_gender, on='Gender')
churned_rates_gender['Churn rate'] = round(churned_rates_gender['Churned'] / churned_rates_gender['Total'] * 100, 2)
churned_rates_gender
```

```python
sns.barplot(
    data = churned_rates_gender,
    x = 'Gender',
    y = 'Total',
    color = 'green'
)

sns.barplot(
    data = churned_rates_gender,
    x = 'Gender',
    y = 'Churned',
    color = 'red'
)

plt.title("Gender")
plt.show()
```

The chart above illustrate the distribution of customers and churned customers based on gender. 
It is observed that the total number of customers are also evenly distributed by gender. and similarly, its shown that reveals that the churn rate is also the same for both genders. 

These findings suggest that customer churn is also not significantly influenced by gender.

#### <a id='toc1_4_2_2_'></a>[**Distribution of customers and customer churn by Contract type and Paperless biling**](#toc0_)

```python
agg_data_contract = data.groupby(['Contract']).size().reset_index(name='Total')
agg_churned_contract = churned.groupby(['Contract']).size().reset_index(name='Churned')

churned_rates_contract = pd.merge(left=agg_data_contract, right=agg_churned_contract, on=['Contract'])
churned_rates_contract['Churn rate'] = round(churned_rates_contract['Churned'] / churned_rates_contract['Total'] * 100, 2)
churned_rates_contract
```

```python
sns.barplot(
    data = agg_data_contract,
    x = 'Contract',
    y = 'Total',
    color = 'green'
)

sns.barplot(
    data = agg_churned_contract,
    x = 'Contract',
    y = 'Churned',
    color = 'red'
)

plt.title("Contract")
plt.show()
```

The chart illustrates the distribution of customers and churned customers based on their contract type. 
The customers are divided into three main groups:
- Month-to-month: The customer pays for the service on a monthly basis without a long-term commitment.
- One year: The customer commits to using the service for one year and typically pays monthly or annually.
- Two year: The customer agrees to use the service for two years, often with a discounted rate.

The analysis reveals that the majority of customers prefer the month-to-month contract over the one-year and two-year contracts.
However, the data also indicates that customers with month-to-month contracts have a significantly higher churn rate compared to those with longer-term contracts.

These findings suggest that contract type plays a significant role in customer retention, likely due to the stronger commitment associated with long-term contracts.

These findings suggest that while paperless billing may have some influence on churn, contract type is a more significant factor in determining customer churn.

## <a id='toc1_6_'></a>[**Key Insights, Recommendations, and Conclusions**](#toc0_)
### <a id='toc1_6_1_'></a>[**Key Insights**](#toc0_)
1. **Customer Characteristics and Churn**:  
   - Senior citizens, customers without dependents, and those with month-to-month contracts have significantly higher churn rates.  
   - Long-term contracts and having tech support are associated with lower churn rates.

2. **Churn by Payment Method**:  
   - Customers using **electronic checks** exhibit the highest churn (~50%), likely due to the prevalence of month-to-month contracts among them.  

3. **Churn by Tenure**:  
   - Churn is highest among customers with shorter tenures, decreasing as tenure increases.

4. **Revenue Impact**:  
   - Churned customers had an average unrecovered value of **$2,617.62**, leading to a **$4.89 million revenue loss**.  
   - While churn rates are comparable across CLTV segments, the total number of churned high-value customers is higher due to their larger base.

5. **Service Preferences and Churn**:  
   - Customers with **fiber optic internet** have higher churn rates than DSL users or those without internet service.  
   - Streaming services (TV or movies) have minimal impact on churn.

6. **Satisfaction and Support**:  
   - Dissatisfaction with service, lack of support, and competitor offerings are the primary reasons for churn.

### <a id='toc1_6_2_'></a>[**Conclusions**](#toc0_)
- **High churn among short-tenure, senior citizens, and electronic check users** highlights vulnerabilities in customer retention strategies.  
- The **CLTV analysis underscores the revenue impact of churn**, emphasizing the importance of retaining high-value customers.  
- Long-term contracts and tech support services are **key retention levers**, with a noticeable influence on reducing churn rates.
- Demography has little to no impact on customer churn

### <a id='toc1_6_3_'></a>[**Recommendations**](#toc0_)
1. **Enhance Retention for High-Churn Groups**:  
   - Develop targeted retention campaigns for **senior citizens** and **short-tenure customers**, such as personalized offers and loyalty programs.  

2. **Promote Long-Term Contracts**:  
   - Incentivize customers to opt for longer contracts by offering discounts, better deals, or value-added services.

3. **Optimize Support Services**:  
   - Improve **tech support availability and quality** to address dissatisfaction and increase retention.  

4. **Address Issues for Electronic Check Users**:  
   - Identify pain points specific to electronic check users (e.g., contract terms) and offer tailored solutions.  

5. **Focus on Competitor Analysis**:  
   - Benchmark services and pricing against competitors to address customer concerns about **better offerings elsewhere**.  

6. **Monitor High-Value Customers**:  
   - Implement proactive retention strategies for high-value customers, such as priority support and exclusive benefits, to minimize revenue loss.  

7. **Leverage Tenure Insights**:  
   - Introduce programs encouraging early-stage customers to stay longer, such as welcome bonuses or onboarding support.
